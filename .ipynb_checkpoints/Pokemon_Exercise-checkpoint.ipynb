```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import seaborn as sns
```

```python
pokemon_df = pd.read_csv('Pokemon.csv')
```

```python
pokemon_df.head()
```

```python
pokemon_hp = pokemon_df['HP']
pokemon_hp
```

```python
hp_bins=[]
hp=0
while hp <= 260:
    hp_bins.append(hp)
    hp+=10
hp_plot = pokemon_hp.plot(kind='hist',xticks=hp_bins,figsize=(12,6),color='green')
hp_plot.set_xlabel('HP')
```

```python
sns.distplot(pokemon_hp,bins=10,hist=True)
```

```python
pokemon_name_hp = pokemon_df[['Name','HP']]
pokemon_name_hp
```

```python
hp_plot2 = pokemon_name_hp.plot(kind='hist',xticks=hp_bins,figsize=(12,6),color='green')
hp_plot2.set_xlabel('HP')
```

```python
pokemon_name_hp.sort_values(by=['HP'],ascending=False).head(10)
```

```python
pokemon_name_hp[pokemon_name_hp.HP>=100].sort_values(by=['HP'],ascending=False)
```

```python
hp_counts = pokemon_name_hp.groupby('HP').count()
hp_counts.rename(columns={'Name':'Count'})
hp_counts
```

```python
name_hp_pivot_table = pokemon_df.pivot_table(index='Name',values='HP')
name_hp_pivot_table
```

```python
hp_plot3 = name_hp_pivot_table.plot(kind='hist',xticks=hp_bins,figsize=(12,6),color='green')
hp_plot3.set_xlabel('HP')
```

```python
hp_counts2 = pokemon_name_hp.groupby('HP').count()
hp_counts2.rename(columns={'Name':'Count'})
hp_counts2
```
