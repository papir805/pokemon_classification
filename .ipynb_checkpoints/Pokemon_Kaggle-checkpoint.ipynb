```python
# Use left join bc we want to preserve all 721 Pokemon numbers after the join.  winners only has pokemon numbers that were used in combat, some of the 721 were never used and aren't found in the sample.  When we do the left join though, any Pokemon numbers in the left df (pokemon) that aren't found in the right df (winners), they'll get a NaN value, so we fill those in at the end.  We also drop_duplicate numbers, keeping the last, so we only focus on highest evolutions that would typically be used in combat.
winners = combats_join['Winner'].value_counts()
pkmn_join = pd.merge(pkmn, winners, how='left', left_on='Number', right_index=True, validate='m:1')
pkmn_join.rename(mapper={'Winner':'Wins'}, axis=1, inplace=True)
pkmn_join.drop_duplicates(subset='Number', keep='last', inplace=True)
pkmn_join['Wins'].fillna(value=0, inplace=True)
pkmn_join.sort_values('Wins', ascending=False)
```

```python
# Get proportions of all pokemon in the combats df
pkmn_join_type_counts = pd.Series(np.zeros(len(pkmn_join_types)), index=pkmn_join_types)
pkmn_join_type_counts = pkmn_join_type_counts.add(pkmn_join['Type 1'].value_counts(), fill_value=0)
pkmn_join_type_counts = pkmn_join_type_counts.add(pkmn_join['Type 2'].value_counts(), fill_value=0)
pkmn_join_type_props = pkmn_join_type_counts / sum(pkmn_join_type_counts)
pkmn_join_type_props
```

```python
def get_zero_series(series):
    "Initilizes a series object of zeros with same length and indices as the series being passed in as the argument"
    temp_series = pd.Series(data=np.zeros(len(series)), index=series)
    return temp_series
```

```python
def get_pop_prop(df):
    pop_type_counts = get_zero_series(pkmn_join_types)
    type1_counts = df['Type 1'].value_counts()
    type2_counts = df['Type 2'].value_counts()
    pop_type_counts = pop_type_counts.add(type1_counts, fill_value=0).add(type2_counts, fill_value=0)
    pop_props = pop_type_counts / sum(pop_type_counts)
    return pop_props
```

```python
def get_tvd(sample_props, pop_props):
    tvd = 0.5 * sum(abs(sample_props - pop_props))
    return tvd
```

```python
def get_topn_props(df, label, n):
    """Calculutes the proportions of top n most winning pokemon by type from a bootsrapped sample."""
    # df_copy = get_bootstrap_df(df, label)
    top_n_counts = get_zero_series(pkmn_join_types)
    df_sorted = df.sort_values(label, ascending=False)
    top_n_type1 = df_sorted["Type 1"][0:n].value_counts()
    top_n_type2 = df_sorted["Type 2"][0:n].value_counts()
    top_n_counts = top_n_counts.add(
        top_n_type1, fill_value=0
    ).add(top_n_type2, fill_value=0)
    top_n_props = top_n_counts / sum(top_n_counts)
    return top_n_props
```

```python
def get_bootstrap_df(df, label, random_state = None):
    """Create bootstrap sample to randomly permute the label column of df, store and return results as df deep copy"""
    sample_label_data = df.sample(df.shape[0], replace=True, random_state = random_state)[label]
    df_copy = df.copy(deep=True)
    df_copy[label] = sample_label_data.values
    return df_copy
```

```python
def get_sample_test_statistic(df, label, n):
    bootstrap_df = get_bootstrap_df(df, label)
    pop_props = get_pop_prop(df)
    bootstrap_props = get_topn_props(bootstrap_df, label, n)
    sample_test_stat = get_tvd(bootstrap_props, pop_props)
    return sample_test_stat
```

```python
def run_simulations(df, label, num_simulations, n=1,):
    #test_statistic = get_test_statistic(df, label, n)
    test_stats = np.array([])
    for i in np.arange(num_simulations):
        #sample_props = get_sample_props(df, label, n)
        sample_tvd = get_sample_test_statistic(df, label, n)
        test_stats = np.append(test_stats, sample_tvd)
    p_value = sum(test_stats>0) / num_simulations  ## Need to more clearly define hypothesis in order to correctly calculate p_value
    return test_stats, p_value
```

```python
def calc_vif(X):
    """
    X: A pandas DataFrame object of numerical independent variables to be used in regression,
    Calculates the variance inflation factor of each independent variable in X
    against all of the other independent variables in X"""
    vif = pd.DataFrame()
    vif['Variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return(vif)
```
